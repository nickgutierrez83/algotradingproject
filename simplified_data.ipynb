```python
'''
Convert date column to datetime format
Select stock price data from 2000 onwards
Set ticker and date as index
Keep only adjusted open, low, high, close, and volume and rename by removing the adj_ prefix
Store in HDF5 format
'''
```

```python
import pandas as pd
```

```python
data = pd.read_csv('prices/us_stocks.csv')
```

```python
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.info()
```

```python
df = data.set_index(['date'])
df.sort_index(inplace=True, ascending=True)
df = data.loc['2020-1-1':]
df.tail()
```

```python
df.head()
```

```python
df = df.drop(['open', 'high', 'low', 'close', 'volume', 'ex-dividend', 'split_ratio'], axis=1)
df.head()
```

```python
df = df.rename(columns={"adj_open": "open", "adj_high": "high", "adj_low": "low", "adj_close": "close", "adj_volume": "volume"})
df.head()
```

```python
h5File = "fromdf.h5"
df.to_hdf(h5File, "data_as_hdf")
```

```python
df1 = pd.read_hdf(h5File, "data_as_hdf")
df1.head()
```
